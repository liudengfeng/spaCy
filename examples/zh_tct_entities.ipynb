```python
import spacy
from spacy.lang.zh import Chinese
from spacy.lang.zh.texsmart import TextSmart
```

```python
nlp = Chinese()
ts = TextSmart() # 默认使用在线模式
```

```python
text = "伟大领袖毛泽东2020年6月12日告诉我们，北京市丰台区新增2例新冠肺炎确诊病例。"
```

# 分词

```python
doc = nlp(text)
[t.text for t in doc]
```

# 命名实体

```python
def tct_entities(doc):
    """重写命名实体"""
    new_ents = []
    spans = ts.get_entities(text)
    for en in spans:
        new_ent = doc.char_span(en[0], en[1], en[2])
        new_ents.append(new_ent)
    doc.ents = new_ents
    return doc
```

```python
# Add the component
nlp.add_pipe(tct_entities, name='tct_ner')

doc = nlp(text)
```

# 显示

```python
from spacy import displacy
displacy.render(doc, style="ent")
```
